# NumPyの基礎

```python
import numpy as np
```

## numpy.ndarray(N-dimentional array)
多次元配列の定義

```python
array1d = np.array([1, 2, 3, 4])
array2d = np.array([[11, 12, 13], [21, 22, 23]])
```

配列の形状

```python
print(array1d.shape)
print(array2d.shape)

# 配列のデータ型
a = np.array([1, 2, 3])
print(a.dtype)
```

## 配列の生成

```python
a = np.arange(1, 10, 2)
print(a)

b = np.linspace(1, 10, 3)
print(b)
```

```python
# 適当な形状の配列を生成(3 x 4 x 5の3次元配列)
a = np.random.rand(3, 4, 5)
print(a)
print(a.shape)
```

## 配列の操作

### スライシング
配列の一部を取り出す。

```python
a = np.array([1, 2, 3, 4])
print(a[1:3]) # 2番目から3番目の要素を取り出す
```

### 形状変更
要素数を変えない範囲で形状を変更する。

```python
array23 = np.array([[1, 2, 3], [4, 5, 6]])
print(array23.shape)

# 2行3列の配列を3行2列の配列に変形する
array32 = array23.reshape(3, 2)
print(array32.shape)

# 2行3列の配列を1行6列の配列に変形する
array16 = array23.reshape(1, 6)
print(array16.shape)
```

### 配列の結合

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
ab = np.concatenate([a, b])
print(ab)
```

## 配列の演算

通常は要素数が同じ配列同士で演算を行う。

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b) # 要素ごとの和
print(a * b) # 要素ごとの積
```

### ブロードキャスト
異なる形状の配列同士で演算を行う場合、ブロードキャストにより自動的に形状が合わせられる。

```python
a = np.array([1, 2, 3])
b = np.array([[1, 2, 3], [4, 5, 6]])
print(a.shape)
print(b.shape)

print(a + b) # ブロードキャストにより、aがbの形状に合わせて自動的に拡張される
```

```python
a = np.array([1, 2, 3])
b = np.array([[1], [2], [3]])
print(a.shape)
print(b.shape)

print(a + b) # ブロードキャストにより、aがbの形状に合わせて自動的に拡張される
```

### 次元を追加して計算を行うパターン

```python
a = np.random.rand(2, 2, 3)
b = np.random.rand(2, 3)
print(a.shape)
print(b.shape)

# bの先頭に要素1の次元が追加され(1, 2, 3)という形状に変換された後に計算が行われる
c = a + b 
print(c.shape)
```

次元追加のイメージ

```python
b = np.random.rand(2, 3)
print(b.shape)
print(b)
# bの先頭に1次元追加
b_expanded = b[np.newaxis, :, :]
print(b_expanded.shape)
print(b_expanded)
```

## 線形代数

内積

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.dot(a, b)) # 内積
```

行列積(行列同士の積)

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a.shape)
b = np.array([[7], [8], [9]])
print(b.shape)
c = np.matmul(a, b) # 行列積
print(c)
```

特定の次元に関してのみ和を計算

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
a_sum1 = np.sum(a, axis=0)
print(a_sum1) # 行方向に和を計算

a_sum2 = np.sum(a, axis=1)
print(a_sum2) # 列方向に和を計算
```
